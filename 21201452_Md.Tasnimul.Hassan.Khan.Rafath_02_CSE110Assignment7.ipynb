```python
# Task 1:

my_list = [10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29]

for iteration in range(0, len(my_list)-1):
    
    for index in range(0, len(my_list)-iteration-1):
        
        if (my_list[index] > my_list[index+1]):
            
            temp = my_list[index]
            
            my_list[index] = my_list[index+1]
            
            my_list[index+1] = temp
            
print(my_list)
```

```python
# Task 4:

sitting_list = [10,30,20,70,11,15,22,16,58,100,12,56,70,80]

even_index_list = []

odd_index_list = []

for index in range(len(sitting_list)):
    
    if (index % 2 == 0):
        
        even_index_list.append(sitting_list[index])
        
    else:
        
        odd_index_list.append(sitting_list[index])
        
even_index_list.sort()

odd_index_list.sort(reverse = True)

final_list = []

for index in range(len(even_index_list)):
    
    final_list.append(even_index_list[index])
    
    final_list.append(odd_index_list[index])
    
print(final_list)
```

```python
# Task 5:

def custom_sorted(names, courses):
    
    for iteration in range(0, len(courses)-1):
        
        for index in range(0, len(courses)-iteration-1):
            
            if (courses[index] < courses[index+1]):
                
                temp = courses[index]
                
                temp_name = names[index]
            
                courses[index] = courses[index+1]
                
                names[index] = names[index+1]
            
                courses[index+1] = temp
                
                names[index+1] = temp_name
                
    for name in names:
        
        print(name)
    
lst = [ ["Alan", 95, 87, 91], ["Turing", 92, 90, 83], ["Elon", 87, 92, 80], ["Musk", 85, 94, 90] ]

names = []

CSE110 = []

PHY111 = []

MAT110 = []

for data in lst:
    
    names.append(data[0])
    
    CSE110.append(data[1])
    
    PHY111.append(data[2])
    
    MAT110.append(data[3])
    
user_input = input("Please enter the course name: ")
    
if (user_input == "CSE110"):
    
    custom_sorted(names, CSE110)
    
elif (user_input == "PHY111"):
    
    custom_sorted(names, PHY111)
    
elif (user_input == "MAT110"):
    
    custom_sorted(names, MAT110)
    
else:
    
    print("Wrong Input!")
```

```python
# Task 7:

list_one = input("Please enter the first list: ").split(", ")

list_two = input("Please enter the second list: ").split(", ")

final_list = []

for num in list_one:
    
    final_list.append(int(num))
    
for num in list_two:
    
    final_list.append(int(num))
    
final_list.sort()

print("Sorted list =", final_list)

if (len(final_list) % 2 == 0):
    
    index = int(len(final_list) / 2)
    
    median = (final_list[index] + final_list[index-1]) / 2
    
else:
    
    index = len(final_list) // 2
    
    median = final_list[index]
    
print("Median =", median)
```

```python
# Task 9:

import math

points = [(1,7), (4,5), (-1,7), (-2,0), (1,1), (5,-1)]

dic = {}

for point in points:
    
    x = point[0]
    
    y = point[1]
    
    total = x**2 + y**2
    
    distance = math.sqrt(total)
    
    dic[distance] = point
    
    min_distance = None
    
    close_point = None
    
for key,value in dic.items():
    
    if (min_distance == None):
        
        min_distance = key
        
        close_point = value
        
    else:
        
        if (key < min_distance):
            
            min_distance = key
        
            close_point = value
            
print("Minimum distance =", min_distance)

print("Here the closest point is", close_point, "which has a distance of", min_distance, "from the origin.")
```
